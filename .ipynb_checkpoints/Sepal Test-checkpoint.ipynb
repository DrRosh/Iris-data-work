```python
from sklearn import datasets
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm as CM

%matplotlib inline
```

```python
iris = datasets.load_iris()
```

```python
type(iris)
```

```python
dir(iris)
```

```python
petals = pd.DataFrame(np.c_[iris["data"], iris["target"]], columns= iris["feature_names"]+["class"])
```

```python
petals.head()
```

```python
petals.describe()
```

```python
plt.scatter(petals.iloc[:,0], petals.iloc[:,1], c=petals["class"])
plt.plot()
```

```python
plt.scatter(petals.iloc[:,2], petals.iloc[:,3], c=petals["class"])
plt.plot()
```

The length and width of petals seem to be very highly correlated so I will only take one of them into my model.

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(petals.iloc[:,0], petals.iloc[:,1], petals.iloc[:,2], c=petals["class"])
plt.show()
```

```python
flowers = petals.drop(columns = petals.columns[3])
sepals = flowers.drop(columns = flowers.columns[2])
```

### K-means
I will start by looking at a basic k-means classification.

```python
kmeans = KMeans(n_clusters = 3, random_state=0).fit(sepals.drop(columns="class"))
```

```python
kmeans.labels_
```

```python
plt.scatter(sepals.iloc[:,0], sepals.iloc[:,1], c=kmeans.labels_)
plt.plot()
```

```python
plt.scatter(petals.iloc[:,0], petals.iloc[:,1], c=petals["class"])
plt.plot()
```

```python
kmeans = KMeans(n_clusters = 3, random_state=0).fit(flowers.drop(columns="class"))
```

```python
plt.scatter(petals.iloc[:,0], petals.iloc[:,1], c=kmeans.labels_)
plt.plot()
```
